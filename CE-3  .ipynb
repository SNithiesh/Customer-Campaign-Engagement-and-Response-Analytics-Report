# Skewness

```python
def skewness(skew_value):
    if -0.5 <= skew_value <= 0.5:
        print("Approximately Normal")
    elif (-1 <= skew_value < -0.5) or (0.5 < skew_value <= 1):
        print("Moderately Skewed")
    else:
        print("Highly Skewed")
```

## Outlier Treatment  And Capping


```python
features2 = [
    'Year_Birth',
    'Recency',
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds',
    'NumDealsPurchases',
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases',
    'NumWebVisitsMonth'
]
```

```python
def iqr_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower) | (data[column] > upper)]
    
    return lower,upper,outliers
```

```python
for col in features2:
    
    lower, upper, outliers = iqr_outliers(data,col)

    print("\nColumn:",col)
    print("Lower:",round(lower,2))
    print("Upper:",round(upper,2))
    print("Outliers:",len(outliers))
```

```python
for col in features2:

    lower, upper, _ = iqr_outliers(data,col)

    data[col]=data[col].clip(
        lower,
        upper
    )
```

```python
for col in features2:
    
    lower, upper, outliers = iqr_outliers(data,col)

    print("\nColumn:",col)
    print("Lower:",round(lower,2))
    print("Upper:",round(upper,2))
    print("Outliers:",len(outliers))
```

### Outliers were capped using lower and upper IQR limits instead of deleting records, preserving customer information while reducing the influence of extreme values.

## Features Selection

```python
X = data.drop(
    'Response',
    axis=1
)

X = X.select_dtypes(
    include=['int64','float64','bool']
)

X = X.astype(float)

y = data['Response']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# convert back to dataframe
import pandas as pd

X_scaled = pd.DataFrame(
    X_scaled,
    columns=X.columns
)
```

### Forward Selection

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SequentialFeatureSelector

model = LogisticRegression(
    max_iter=20000,
    solver='liblinear'
)

forward = SequentialFeatureSelector(
    model,
    n_features_to_select=5,
    direction='forward'
)

forward.fit(X_scaled,y)

forward_features = X_scaled.columns[
    forward.get_support()
]

print("Forward")
print(forward_features)
```

#### Forward Selection identified variables that progressively improved predictive performance. AcceptedCmp1, AcceptedCmp3, and AcceptedCmp5 indicate that a customer's response to previous marketing campaigns strongly influences future responses. Recency suggests the time since a customer's last purchase impacts engagement behavior. Teenhome indicates household composition also contributes to customer response prediction.

### Backward Elimination

```python
backward = SequentialFeatureSelector(
    model,
    n_features_to_select=5,
    direction='backward'
)

backward.fit(X_scaled,y)

backward_features = X_scaled.columns[
    backward.get_support()
]

print("Backward")
print(backward_features)
```

#### Backward Elimination retained variables that remained significant even after starting with the full feature set. Family_Size replaced Teenhome, indicating that broader household structure provides more meaningful information than a single household attribute. Recency and campaign acceptance variables continued to appear, showing stable predictive relevance.

## Lasso 

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

lasso = Lasso(alpha=0.01)

lasso.fit(X_scaled,y)

lasso_model = SelectFromModel(
    lasso,
    prefit=True
)

lasso_features = X_scaled.columns[
    lasso_model.get_support()
]

print("\nLasso")
print(lasso_features)
```

#### Lasso retained variables with stronger predictive contribution while reducing less useful coefficients. Multiple purchase categories (MntFruits, MntMeatProducts, MntGoldProds) and purchase channels (NumCatalogPurchases, NumStorePurchases) were selected, suggesting customer buying patterns and purchasing preferences strongly influence response behavior.

## Information Gain

```python
from sklearn.feature_selection import mutual_info_classif
import pandas as pd

scores = mutual_info_classif(
    X_scaled,
    y
)

ig = pd.DataFrame({
    'Feature':X_scaled.columns,
    'Score':scores
})

print("\nInformation Gain")

print(
ig.sort_values(
'Score',
ascending=False
).head(10)
)
```

#### MntMeatProducts received the highest Information Gain score, meaning this feature reduced uncertainty about customer response more than others. AcceptedCmp5, Income, and Total_Spending also showed meaningful contribution. Higher scores indicate greater influence on prediction.

## Fisher Score 

```python
from skfeature.function.similarity_based import fisher_score

ranks = fisher_score.fisher_score(
X_scaled.values,
y.values
)

fisher = pd.DataFrame({
'Feature':X_scaled.columns,
'Score':ranks
})

print("\nFisher Score")

print(
fisher.sort_values(
'Score',
ascending=False
).head(10)
)
```

#### Campaign acceptance variables achieved high Fisher scores, showing strong ability to separate customer classes. NumStorePurchases and NumDealsPurchases also contributed to class separation, indicating differences in purchasing behavior among customer groups

```python
from collections import Counter

# Forward and Backward
forward_set = set(forward_features)
backward_set = set(backward_features)

# Lasso
lasso_set = set(lasso_features)

# Top 5 Fisher features
fisher_features = fisher.sort_values(
    'Score',
    ascending=False
)['Feature'].head(5)

fisher_set = set(fisher_features)

# Top 5 Information Gain features
ig_features = ig.sort_values(
    'Score',
    ascending=False
)['Feature'].head(5)

ig_set = set(ig_features)

# Combine all selected features
all_features = (
    list(forward_set)
    + list(backward_set)
    + list(lasso_set)
    + list(fisher_set)
    + list(ig_set)
)

# Count frequency
feature_count = Counter(all_features)

print("\nFeature Counts:")
for feature, count in feature_count.items():
    print(feature, ":", count)

# Final important features
important = [
    feature
    for feature, count in feature_count.items()
    if count >= 3
]

print("\nSelected by at least 3 methods:")
print(important)
```
